```python
import pandas as pd
import numpy as np
from sentence_transformers import util
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('Sahajtomar/German-semantic')
```

```python
df=pd.read_csv("번역 결과.csv",encoding="utf-8-sig")

result=pd.DataFrame({"원문":[],
                    "번역본":[],
                    "유사도":[]})
df
```

```python
total=0.0
for idx in range(0,1000):
    item=df.loc[idx]
    original=item[-1]
    trans=item[-2]
    embeddings1 = model.encode(original, convert_to_tensor=True)
    embeddings2 = model.encode(trans, convert_to_tensor=True)
    cosine_scores = util.pytorch_cos_sim(embeddings1, embeddings2)
    now=pd.DataFrame({"원문":[original],
                    "번역본":[trans],
                    "유사도":[float(cosine_scores[0][0])]})
    result=pd.concat([result,now])
    total+=float(cosine_scores[0][0])
total/=1000
print(total)
```
